```python
import numpy as np
import matplotlib.pyplot as plt
plt.style.use('./deeplearning.mplstyle')
```

```python
X_train = np.array([1.0,2.0])
y_train = np.array([300,500])
print(f"X_train = {X_train}")
print(f"y_train = {y_train}")
```

```python
print(f"x_train = {X_train.shape}")
m = X_train.shape[0]
print(f"y_train = {m}")
```

```python
len(X_train)
```

```python
i=0
x_i = X_train[0]
y_i = y_train[0]
print(f"x^({i}),y^({i})=({x_i},{y_i})")
```

```python
plt.scatter(X_train,y_train,marker='x',c='r')
plt.title('House price pridiction')
plt.ylabel('price in dollar(in 1000)')
plt.xlabel('size in sq feet(in 1000)')
```

```python
w=100
b=100
```

```python
def computeOutputModel(x,w,b):
    m=X_train.shape[0]
    f_wb = np.zeros(m)
    for i in range(m):
        f_wb[i] = w*x[i]+b
        
    return f_wb
```

```python
temp_f_wb = computeOutputModel(X_train,w,b)
plt.scatter(X_train,y_train,c='r',marker="*",label="actual value")
plt.plot(X_train,temp_f_wb,c='b',label="our prediction")
plt.title('price prediction')
plt.ylabel('price in dollar(in 1000)')
plt.xlabel('size in sq feet(in 1000)')
plt.legend()
plt.show()

```
